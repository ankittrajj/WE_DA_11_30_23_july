```python
# mat plot lib
# mathematical plotting lib.

# Line plot
# Histogram
# Barchart
# Pie chrt
# subplot

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Graphs 
no_of_classes = np.array([1,2,3,4])
no_of_students = np.array([34,67,21,90])

plt.plot(no_of_classes,no_of_students)
```

```python
no_of_classes = np.array([1,2,3,4])
no_of_students = np.array([34,67,21,90])
plt.xlabel('Number of classes')
plt.ylabel('Number of students')
plt.title('Classes vs Students Graph')

plt.plot(no_of_classes,no_of_students)
```

```python
# Wap to make a graph for temp vs city.
city = ['A','B','C','D']
temp = [23,34,15,37]
plt.plot(city,temp,color = 'r')
```

```python
#wap to make a graph for 2cities vs temp.

city1 = ['A','B','C']
city2 = ['X','Y','Z']

temp1 = [34,23,9]
temp2 = [5,23,45]

plt.plot(city1,temp1,city2,temp2)
```
