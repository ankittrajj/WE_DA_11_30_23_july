```python
# merge,concate,append
x1= {
    "name":["aman","jai"],
    "age":[45,87],
    "marks":[450,987],
    "hobby":'dance'
}
x2={
    "name":[ "aman","johnny"],
    "age": [23,89],
    "marks":[870,980]

}
```

```python
x1
```

```python
x2
```

```python
# change the dict into data frames.
df1= pd.DataFrame(x1)
```

```python
df1
```

```python
df2 = pd.DataFrame(x2)
```

```python
df2
```

```python
# merge
s = pd.merge(df1,df2,how = 'left')
```

```python
s
```

```python
s = pd.merge(df1,df2,how = 'right')
```

```python
s
```

```python
s = pd.merge(df1,df2,how = 'inner')
```

```python
s
```

```python
s = pd.merge(df1,df2,how = 'outer')
```
